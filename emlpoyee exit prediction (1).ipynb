# Data preprocessing

```python
#missing value
```

```python
df.isnull().sum()  # no missing values
```

```python
# feature encoding
```

```python
df_ohe=pd.get_dummies(df[["sales","salary"]])
```

```python
df=pd.concat([df,df_ohe],axis=1)
```

```python
df.drop(["sales","salary"],axis=1,inplace=True)  
```

```python
df
```

```python
# all columns are converted into the numerical columns
```

```python
# feature scaling
```

```python
df.describe()
```

```python
# we need to scale the "average_montly_hours" feature

from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
mm
```

```python
df["average_montly_hours"]=mm.fit_transform(df[["average_montly_hours"]])
df
```

# divding and slitting x and y

```python
x=df.drop(["left"],axis=1)
y=df["left"]
```

```python
#spltting
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# apply Kneighbourclassifer

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
